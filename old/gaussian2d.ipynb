```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib widget
```

```python
def gauss1d(x,mu,sigp,sigm):
    if x > mu:
        sig = sigp
    else:
        sig = sigm
    return np.exp(-0.5 * (x-mu)**2/sig**2)
```

```python
fig,ax = plt.subplots(1,1)

mu  = 1.0
sigp = 0.5
sigm = 1.0

xarr = np.linspace(-5, 5, 200)
yarr = [gauss1d(x, mu, sigp, sigm) for x in xarr]

ax.plot(xarr, yarr)

plt.show()
```

```python
fig,ax = plt.subplots(1,1)

x_par = [1.0, 0.5, 1.0]
y_par = [0.5, 1.0, 3.0]

x_inp = np.linspace(-5, 5, 200)
y_inp = np.linspace(-5, 5, 200)

x_arr, y_arr, z_arr = [], [], []
for x in x_inp:
    for y in y_inp:
        x_arr.append(x)
        y_arr.append(y)

        z = gauss1d(x, x_par[0], x_par[1], x_par[2]) * gauss1d(y, y_par[0], y_par[1], y_par[2])
        z_arr.append(z)

c = 'white'
ax.axvline(x_par[0], c=c)
ax.axvline(x_par[0]+x_par[1], c=c, ls='dashed')
ax.axvline(x_par[0]-x_par[2], c=c, ls='dashed')

ax.axhline(y_par[0], c=c)
ax.axhline(y_par[0]+y_par[1], c=c, ls='dashed')
ax.axhline(y_par[0]-y_par[2], c=c, ls='dashed')

ax.tricontourf(x_arr, y_arr, z_arr)

plt.show()
```
