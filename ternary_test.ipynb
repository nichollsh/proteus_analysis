```python
import matplotlib as mpl

mpl.rcParams.update({'font.size': 12})
mpl.rcParams['axes.formatter.useoffset'] = False

import matplotlib.pyplot as plt
from ternary.helpers import simplex_iterator
import ternary, math
import numpy as np
```

## Scatter plot

```python
scale = 100.0
points = []
for i in range(40):
    rnd = np.random.sample(3)
    rnd = rnd/np.sum(rnd) * scale
    points.append(rnd)

### Scatter Plot
fig,ax = plt.subplots()
tfg, tax = ternary.figure(ax, scale=scale)
tax.boundary(linewidth=2.0)
tax.scatter(points, marker='s', color='red')
tax.ticks(axis='lbr', linewidth=1, multiple=20.0)

ax.set_axis_off()

tax.show()
```

## Heat map

```python

scale = 10

def generate_heatmap_data():
    d = dict()
    for (i, j, k) in simplex_iterator(scale):
        d[(i, j, k)] = np.random.sample(1) * (i+j)/scale
    return d

data = generate_heatmap_data()

fontsize=12

fig,ax = plt.subplots()
tfg, tax = ternary.figure(ax=ax, scale=scale)
tax.heatmap(data, style="triangular", cmap='cividis')
tax.boundary()
tax.ticks(axis='lbr', linewidth=1, multiple=scale/5,  offset=0.025)
tax.left_axis_label(  "H2O VMR", offset=0.13)
tax.right_axis_label( "CO2 VMR", offset=0.15)
tax.bottom_axis_label("H2 VMR",  offset=0.09)


ax.set_axis_off()

tax.show()
```
