## Histograms and similar plots

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import numpy as np

%config InlineBackend.figure_format='retina'
```

```python
import utils.load_grid as lg
from utils.load_cmaps import *
```

```python
pgrid_dir = "data/scratch/pgrid_earth_gridtest_2"
cases = lg.get_cases(pgrid_dir)
```

```python
# Manual variables
endt = np.array([lg.get_nc_years(c)[-1] for c in cases])
endp = np.array([lg.read_nc(c+"/data/%d_atm.nc" % endt[i])["psurf"] for i,c in enumerate(cases)])
stts = lg.get_statuses(pgrid_dir)

good = [bool(10 <= s <= 19) for s in stts]
```

### Termination year

```python
fig,ax = plt.subplots(1,1)

ax.hist(endt[good], color='black')
ax.set_xlabel("Termination [yr]")
ax.set_ylabel("Count")
ax.yaxis.set_major_formatter(PercentFormatter(len(endt[good])))

plt.show()
```

### Final total surface pressure

```python
fig,ax = plt.subplots(1,1)

ax.hist(endp[good]*1.0e-5, color='black')
ax.set_xlabel("Surface pressure at termination [bar]")
ax.set_ylabel("Count")
ax.yaxis.set_major_formatter(PercentFormatter(len(endp[good])))

plt.show()
```
