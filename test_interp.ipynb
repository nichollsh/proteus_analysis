```python
import utils.load_grid as lg
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [-2, 0, 4, 5, 8, 15]
y = [0.01, 0.1, 10, 100, 200]
z = np.random.sample(len(x)*len(y))
```

```python
xg,yg,X,Y,D = lg.interp_2d(x,y,z, 800, method='linear')

fig,ax = plt.subplots()

ax.pcolormesh(X, Y, D)
ax.scatter(xg, yg, c=z, edgecolor='white', zorder=3)

ax.set_yscale("log")

plt.show()
```
